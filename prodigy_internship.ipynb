```python
import matplotlib.pyplot as plt
import pandas as pd

age_counts = d['Age'].value_counts()  # Calculate age counts

# Plot bar chart
plt.figure(figsize=(5, 4))
plt.bar(age_counts.index, age_counts.values, color='lightblue')
plt.xlabel('AGE')
plt.ylabel('FREQUENCY')
plt.title('DISTRIBUTION OF AGES (Bar Chart)')
plt.xticks(age_counts.index)  # Ensure all age labels are shown
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
